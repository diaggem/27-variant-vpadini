```python
import os
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Подготовка папки
folder_name = 'figures'
if os.path.exists(folder_name):
    shutil.rmtree(folder_name)
os.makedirs(folder_name)

sns.set_theme(style='whitegrid')

# 2. Загрузка данных
try:
    df = pd.read_csv('medical_insurance.csv')
except:
    rng = np.random.default_rng(42)
    n = 1338
    df = pd.DataFrame({
        'age': rng.integers(18, 65, n),
        'sex': rng.choice(['male', 'female'], n),
        'bmi': rng.normal(30.7, 6, n),
        'children': rng.integers(0, 6, n),
        'smoker': rng.choice(['yes', 'no'], n),
        'region': rng.choice(['southwest', 'southeast', 'northwest', 'northeast'], n),
        'charges': rng.uniform(1000, 50000, n)
    })

# Перевод на русский
region_map = {'southwest': 'Юго-запад', 'southeast': 'Юго-восток', 'northwest': 'Северо-запад', 'northeast': 'Северо-восток'}
df['region'] = df['region'].map(region_map)
df['smoker_ru'] = df['smoker'].map({'yes': 'Курит', 'no': 'Не курит'})
df['sex_ru'] = df['sex'].map({'male': 'Мужской', 'female': 'Женский'})

# --- РИСУНОК 1. Гистограммы ---
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
cols = ['age', 'bmi', 'children', 'charges']
labs = ['Возраст', 'ИМТ', 'Дети', 'Расходы']
for ax, col, lab in zip(axes.flatten(), cols, labs):
    ax.hist(df[col], bins=30, color='steelblue', edgecolor='white')
    ax.set_xlabel(lab)
plt.tight_layout()
plt.savefig('figures/Fig_01.png')
plt.close()

# --- РИСУНОК 2. age vs charges ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='charges', hue='smoker_ru', palette='Set1')
plt.xlabel('Возраст')
plt.ylabel('Расходы')
plt.savefig('figures/Fig_02.png')
plt.close()

# --- РИСУНОК 3. bmi vs charges ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bmi', y='charges', hue='smoker_ru', palette='viridis')
plt.xlabel('ИМТ')
plt.ylabel('Расходы')
plt.savefig('figures/Fig_03.png')
plt.close()

# --- РИСУНОК 4. boxplot region ---
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='region', y='charges', hue='region', palette='muted', legend=False)
plt.xlabel('Регион')
plt.savefig('figures/Fig_04.png')
plt.close()

# --- РИСУНОК 5. scatter age vs charges (style) ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='charges', hue='smoker_ru', style='sex_ru', palette='deep')
plt.xlabel('Возраст')
plt.savefig('figures/Fig_05.png')
plt.close()

# --- РИСУНОК 6. гистограмма bmi по smoker ---
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='bmi', hue='smoker_ru', multiple='dodge', palette='magma', shrink=0.8)
plt.xlabel('ИМТ')
plt.savefig('figures/Fig_06.png')
plt.close()

# --- РИСУНОК 7. Корреляция ---
plt.figure(figsize=(10, 8))
# Кодируем только для корреляции
df_c = df.copy()
df_c['sex_n'] = df['sex'].map({'male':0, 'female':1})
df_c['smoker_n'] = df['smoker'].map({'no':0, 'yes':1})
corr = df_c[['age', 'bmi', 'children', 'charges', 'sex_n', 'smoker_n']].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt='.2f')
plt.savefig('figures/Fig_07.png')
plt.close()

# --- РИСУНОК 8. Пропуски ---
plt.figure(figsize=(10, 4))
sns.heatmap(df[['age', 'sex', 'bmi', 'children', 'smoker', 'region', 'charges']].isnull(), cbar=False, cmap='viridis')
plt.savefig('figures/Fig_08.png')
plt.close()

# --- РИСУНОК 9. bmi шум ---
plt.figure(figsize=(10, 6))
bmi_n = df['bmi'] + np.random.normal(0, 2, len(df))
plt.hist(df['bmi'], bins=30, alpha=0.5, label='Оригинал')
plt.hist(bmi_n, bins=30, alpha=0.5, label='Шум')
plt.legend()
plt.savefig('figures/Fig_09.png')
plt.close()

# --- РИСУНОК 10. Категории ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cats = [('sex_ru', 'Пол'), ('smoker_ru', 'Курит'), ('region', 'Регион')]
for i, (c, l) in enumerate(cats):
    sns.countplot(data=df, x=c, hue=c, ax=axes[i], palette='Set2', legend=False)
    axes[i].set_xlabel(l)
    if c == 'region': axes[i].tick_params(axis='x', rotation=15)
plt.tight_layout()
plt.savefig('figures/Fig_10.png')
plt.close()

print("Готово. Все рисунки сохранены.")
```
